```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("HeartDiseases.csv")
df1.head(10)
```

```python
df1.isna().sum()
```

```python
df2 = df1.copy()
df2['glucose'] = df2['glucose'].fillna(df2['glucose'].mean())
df2['BMI'] = df2['BMI'].fillna(df2['BMI'].mean())
df2['totChol'] = df2['totChol'].fillna(df2['totChol'].mean())
```

```python
df2['BPMeds'].unique()
```

```python
df2['BPMeds'] = df2['BPMeds'].fillna(0)
```

```python
df2['cigsPerDay'].unique()
```

```python
df2['TenYearCHD'].unique()
```

```python
df3 = df2.dropna()
```

```python
df3.isna().sum()
```

```python
plt.figure(figsize = (20, 15))
sns.heatmap(df3.corr(),annot=True)
plt.show()
```

```python
sns.pairplot(df3)
plt.show()
```

```python
df4 = df3.drop(['currentSmoker'],axis = 1)
df4.head(3)
```

```python
x = df4.drop(['TenYearCHD'],axis = 1)
y = df4['TenYearCHD']
```

```python
x.head(3)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 10)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(x_train,y_train)
reg.score(x_test,y_test)
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
```

```python
algos = {
    'svm': {
        'model': svm.SVC(),
        'params' : {
            'C': [1,10,100],
            'kernel': ['rbf']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'C': [1,5,10]
        }
    },
    'decision_tree': {
        'model' : DecisionTreeClassifier(),
        'params' : {
            'criterion' : ['gini', 'entropy', 'log_loss'],
            'splitter' : ['best', 'random']
        }
    },
    'KNN' : {
        'model' : KNeighborsClassifier(),
        'params' : {
            'weights' : ['uniform', 'distance'],
            'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute']
        }
    },
    'naive_bias' : {
        'model' : GaussianNB(),
        'params' : {
            
        }
    }
}
```

```python
scores = []
best_estimators = {}
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
for algo_name, config in algos.items():
    gs =  GridSearchCV(config['model'], config['params'], cv=cv)
    gs.fit(x,y)
    scores.append({
        'model': algo_name,
        'best_score': gs.best_score_,
        'best_params': gs.best_params_,
    })
    
    best_estimators[algo_name] = gs.best_estimator_
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
best_estimators
```

```python
best_clf = best_estimators['logistic_regression']
```

```python
best_clf.score(x_test,y_test)
```

```python
predict_values = best_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,predict_values)
cm
```
